```python
import numpy as np
df['Total Spent']=df['Total Spent'].replace('UNKNOWN',np.nan)
df['Total Spent']=df['Total Spent'].replace('ERROR',np.nan)
df['Price Per Unit']=df['Price Per Unit'].replace('UNKNOWN',np.nan)
df['Price Per Unit']=df['Price Per Unit'].replace('ERROR',np.nan)
```

```python
df['Total Spent'].unique()
```

```python
df['Price Per Unit'].unique()
```

```python
df['Total Spent']=df['Total Spent'].astype('float64')
df['Price Per Unit']=df['Price Per Unit'].astype('float64')
```

```python
# filling the rows in the quantity column where their is 0 
mask=(
    df['Total Spent'].notna() &
    df['Price Per Unit'].notna() &
    df['Quantity'].isna()
)

df.loc[mask,'Quantity']=df.loc[mask,'Total Spent'] / df.loc[mask,'Price Per Unit']
```

```python
#counting such rows in the quantity column where values exists for both Price Per Unit Column and Total Spent Column
df.loc[df['Price Per Unit'].notna() & df['Total Spent'].notna() & df['Quantity'].isna(), 'Quantity'].shape[0]
```

```python
mask1=(
    df['Quantity'].notna() &
    df['Price Per Unit'].notna() &
    df['Total Spent'].isna()
)

df.loc[mask1,'Total Spent']=df.loc[mask1,'Quantity']*df.loc[mask1,'Price Per Unit']
```
