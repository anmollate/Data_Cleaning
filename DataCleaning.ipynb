```python
#Replacint The null values in the Quantity=Total Spent/Price Per Unit
mask=(
    df['Total Spent'].notna() &
    df['Price Per Unit'].notna() &
    df['Quantity'].isna()
)

df.loc[mask,'Quantity']=df.loc[mask,'Total Spent'] / df.loc[mask,'Price Per Unit']
```
